```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Step 1: Load the dataset
df = pd.read_csv('E-commerce Customer Behavior - Sheet1 (1).csv')

# Step 2: Convert relevant columns to categorical values (for transaction-style data)
df['Gender'] = 'Gender_' + df['Gender'].astype(str)
df['City'] = 'City_' + df['City'].astype(str)
df['Membership Type'] = 'Membership_' + df['Membership Type'].astype(str)
df['Discount Applied'] = 'Discount_' + df['Discount Applied'].astype(str)
df['Satisfaction Level'] = 'Satisfaction_' + df['Satisfaction Level'].astype(str)

# You can bin numerical columns to convert into categories
df['Age'] = pd.cut(df['Age'], bins=[0, 25, 45, 100], labels=['Age_Young', 'Age_Middle', 'Age_Senior'])
df['Total Spend'] = pd.cut(df['Total Spend'], bins=3, labels=['Spend_Low', 'Spend_Medium', 'Spend_High'])
df['Items Purchased'] = pd.cut(df['Items Purchased'], bins=3, labels=['Items_Few', 'Items_Medium', 'Items_Many'])

# Step 3: Create a list of lists for transactions
transactions = df[['Gender', 'City', 'Membership Type', 'Discount Applied', 
                   'Satisfaction Level', 'Age', 'Total Spend', 'Items Purchased']].values.tolist()

# Step 4: One-hot encode the data
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

# Step 5: Apply Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

# Step 6: Generate association rules
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)

# Step 7: Display results
print("ðŸ“¦ Frequent Itemsets:")
print(frequent_itemsets)

print("\nðŸ”— Association Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
